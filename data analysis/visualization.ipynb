```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#bar plot
sales = np.array([100, 200, 300, 400, 500,420, 300, 500, 600])
months = np.array(['jan','Feb','Mar','Apr','May'])
plt.bar(x=months, height=sales)
plt.show()
```

```python
#bar chart with text
plt.bar(x=months, height=sales)
for i in range(len(sales)):
    plt.text(x=i, y=10, s=sales[i], ha='center', fontdict={'color':'white'})
plt.show()
```

```python
plt.bar(y=months, width=sales)
plt.show()
```

```python
plt.bar(x=months, height=sales, color='green', edgecolor='black', alpha=.5)
```

```python
colors = ['red','green','blue','yellow','orange','purple','gray','brown']
plt.bar(x=months, heigth=sales, color=colors, edgecolor='black', alpha=.5)

```

```python
#box plot
fig = plt.figure(figsize=(20,6))
x = np.random.randint(1,10000,(100,3))
```

```python
#box plot
S1_marks = [95,83,94,95,72]
S2_marks = [45,93,90,65,72]
S3_marks = [75,63,70,55,62]
marks = np.array([S1_marks, S2_marks, S3_marks]).T
marks
```

```python
plt.boxplot(marks, labels=['S1', 'S2', 'S3'], showmeans=True, meanline=True)
plt.show()
```

```python
#stem plot
fig = plt.figure(figsize=(20,6))
x = np.linspace(0, 10, 50)
y = np.sin(x)
plt.stem(x, y, linefmt='r:', markerfmt='r*', basefmt='g')
plt.show()
```

```python
#hexbin plot
fig = plt.figure(figsize=(8,6))
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize= 25, cmap='copper')
plt.colorbar()
plt.show()
```
